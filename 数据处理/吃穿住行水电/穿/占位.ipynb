```python
# 输入数据 && 预处理

begin_year = 2017

eat_2017 = [1252.3, 985.78, 971.53, 1074.71, 533.3,
            245.58, 1315.52, 956.71, 1284.56, 1153.05]
eat_2018 = [1363.75, 1006.5, 1802.3, 1543.81, 1404.54,
            1242.37, 1314.24, 1239.3, 1287.3, 1078.65, 1107, 1317.7]
eat_2019 = [1778.08, 788.9, 1270.1, 1356.8, 1414.5,
            1350.65, 1760.49, 1705.17, 1370.6, 1314, 1468.3, 982.3]
eat_2020 = [736.7, 466.5, 445.5, 714, 744.2, 1402.1,
            1239.1, 622.6, 1570.94, 1277.8, 1289.7, 1255]
eat_2021 = [906.8, 801.4, 1551.3, 1410.2, 1566.28, 1408.34,
            1533.9, 1439.4, 1449.32, 1788.92, 1911.33, 1529.79]
eat_2022 = [3394.23, 1544.23, 1604.7, 1530.6, 1550.2,
            2495.98, 1820.26, 1392.9, 1182.1, 1761.8, 1863.5, 1922.7]
eat_2023 = [2181.8, 2154.33, 2326.97, 2488.58, 2378.96]

# 2017 年 3 月开始的，1、2 月特殊处理，赋值为其他 10 个月的平均值
eat_begin_year = eat_2017
eat_begin_year_avg = round(sum(eat_begin_year)/len(eat_begin_year), 2)
while (len(eat_begin_year) < 12):
    eat_begin_year = [eat_begin_year_avg] + eat_begin_year
print('begin year:', eat_begin_year)
print()

# 今年剩下几个月的值赋为今年的平均值
eat_end_year = eat_2023
eat_end_year_avg = round(sum(eat_end_year)/len(eat_end_year), 2)
while (len(eat_end_year) < 12):
    eat_end_year.append(eat_end_year_avg)
print('this year:', eat_end_year)
print()

eat_data_2d = [
    eat_begin_year,
    eat_2018,
    eat_2019,
    eat_2020,
    eat_2021,
    eat_2022,
    eat_end_year,
]
```

```python
year_total_eat = [int(sum(i)) for i in eat_data_2d]
print("每年总共吃多少元?")
for i, p in enumerate(year_total_eat):
    print(str((begin_year+i))+"年: "+str(int(p)) +"元")
print()


year_avg_eat = []
for i, p in enumerate(eat_data_2d):
    year_avg_eat.append(int(sum(p)/len(p)))
print("每年平均吃多少元?")
for i, p in enumerate(year_avg_eat):
    print(str((begin_year+i))+"年: "+str(int(p))+" 元")
print()

year_avg_eat_growth_rate = [0]
for i in range(1, len(year_avg_eat)):
    growth_rate = round((year_avg_eat[i]-year_avg_eat[i-1])/year_avg_eat[i-1]*100,1)
    year_avg_eat_growth_rate.append(growth_rate)
print("年平均吃的增长率百分比:", year_avg_eat_growth_rate)
print()

eat_data_1d = []
for i in eat_data_2d:
    eat_data_1d += i
# print("2017年3月以来每个月吃:", eat_data_1d); 
# print()

print("一共吃了"+str(int(sum(year_total_eat))) + "元, 平均每年吃 " + str(int(sum(year_total_eat)/len(year_total_eat)))+" 元, 平均每月吃 "+str(int(sum(eat_data_1d)/len(eat_data_1d))) +"元")
print()
month_avg_eat_growth_rate = [0]
for i in range(1, len(eat_data_1d)):
    growth_rate = round((eat_data_1d[i]-eat_data_1d[i-1])/eat_data_1d[i-1]*100, 1)
    month_avg_eat_growth_rate.append(growth_rate)
print("月平均吃的增长率百分比: ", month_avg_eat_growth_rate)
print()
```
