```python
id_file_path = '计算机学院-博导-94人.html'

with open(id_file_path, 'r', encoding='utf-8') as fid:
    lines = fid.readlines()

ids =[]

for line in lines:
    if 'onView(\'' not in line:
        continue
    index = line.find('onView(\'')
    if not line[index+8].isdigit():
        continue
    t1 = line.split('onView(\'')
    t1=t1[1]
    t1 = t1.split('\')')[0]
    ids.append(t1)
print(len(ids))
```

```python
import requests
from bs4 import BeautifulSoup
 
infos = []

def get_info(id):
    url = "http://yjszs.nudt.edu.cn:80/pubweb/homePageList/getView.view?keyId="+id
    
    # 发送HTTP请求
    response = requests.get(url)
    
    # 检查请求是否成功
    if response.status_code == 200:
        # 使用BeautifulSoup解析网页内容
        soup = BeautifulSoup(response.text, 'html.parser')
        # print('soup:', soup)
        # 找到所有的表格
        tables = soup.find_all('table')
        
        # 遍历所有表格
        one_person_info = []
        for table in tables:
            # 找到所有的行
            for row in table.find_all('tr'):
                # 找到所有的列
                cols = row.find_all('td')
                # 打印列中的文本
                for col in cols:
                    if col.text[-1] == '：':
                        continue
                    one_person_info.append(col.text.strip())
                # print()  # 换行
    else:
        print("Failed to retrieve the webpage")
    return one_person_info

for id in ids:
    infos.append(get_info(id))
```

```python
from openpyxl import load_workbook, Workbook
cwb = Workbook()
cws = cwb.active
for i, row in enumerate(infos):
    print(f'第{i+1}人：{row}')
    # la = row[-1]
    # la = la.index('。')
    # print(i, la)
    cws.append(row)

cwb.save("老师信息.xlsx")
```
